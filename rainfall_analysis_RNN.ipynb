```python
#state wise (jun-sept) historical data 
data=dataset1[['SUBDIVISION','YEAR','Jun-Sep']]
data.columns=['SUBDIVISION','YEAR','Js']
data.head()
```

```python

#choosing a particular state data to mimimalize the moving average of rainfall
data=data[data['SUBDIVISION']=='TAMIL NADU' ]
data = data.reset_index(drop=True)
data.head()
```

```python
#filetring the data with box plot analysis
import seaborn as sns
sns.boxplot(x=data['Js'])
```

```python
#applying upper quartile limt and lower quartile limit for the data to remove the outliers

data = data[np.isfinite(data['Js'])]
data=data[data.Js>290]
data=data[data.Js <390]
```

```python
#cleaned dataframe
data = data.reset_index(drop=True)
data.head()
```

```python

#data virtualization
fig = plt.figure(figsize=(13,8))
ax = fig.add_subplot(111)
dfg = data.groupby('YEAR').mean()['Js']
dfg.plot('line', title='monthly average  Rainfall in tamil nadu ', fontsize=20,color='g')
plt.ylabel('Overall Rainfall ')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)

ax.yaxis.label.set_fontsize(20)
```

```python

import tensorflow as tf 
from sklearn.preprocessing import MinMaxScaler
```

```python
# numpy array "y" has the list of average rainfall of (jun-sept) from cleaned dataframe
# numpy array "x" has the list of year from cleaned dataframe

y = data['Js']
cols =['YEAR']
x=data['YEAR']
```

```python
#preprocessing with one-hot encoding (scalling values from (-1 to 1)) to reduce the complexity of calculation of gradient descent


from sklearn import preprocessing
scaler_x = preprocessing.MinMaxScaler ( feature_range =( -1, 1))
x = np. array (x).reshape ((len( x) ,len(cols)))
x = scaler_x.fit_transform (x)
x
```

```python
#preprocessing with one-hot encoding (scalling values from (-1 to 1)) to reduce the complexity of calculation of gradient descent
scaler_y = preprocessing. MinMaxScaler ( feature_range =( -1, 1))
y = np.array (y).reshape ((len( y), 1))
y = scaler_y.fit_transform (y)
y
```

```python
#splitting the cleaned dataframe into 80% and 20% , 80% xtrain and ytrain for traning  ,20% x_test and y_test for validation 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python

#obtaining the shape of the one-hot encoded arrays for xtrain and xtest for neural net
x_train = x_train.reshape (x_train. shape + (1,)) 
x_test = x_test.reshape (x_test. shape + (1,))
```

```python
#importing dependencies
from keras.models import Sequential
from keras.layers.core import Dense
from keras.layers.recurrent import LSTM
from keras.layers import  Dropout
```

```python
seed =2019
np.random.seed (seed)
```

```python

pred = Sequential ()
pred.add (LSTM (  2000 , activation = 'relu', input_shape =(len(cols), 1) ))
pred.add(Dense(1000))
pred.add(Dropout(0.2))
pred.add (Dense (output_dim =1, activation = 'sigmoid'))
```

```python

#printing the structure of the lstm model
print(pred.summary())
```

```python
# defining the error metrix to reach the local minima of gradient descent for sequential predicitons

pred.compile (loss ="mean_squared_error" , optimizer = "adam")   
```

```python
#training the lstm net to minimize mean squared error

pred.fit (x_train, y_train, batch_size =10, epochs =20)
```

```python
#validating the trained model with test data

score_train = pred.evaluate (x_train, y_train, batch_size =1)
score_test = pred.evaluate (x_test, y_test, batch_size =1)
```

```python
# 0.30 ,0.35 implies that 30% failure rate for the neural network to obtain the desired global minima

print (" in train MSE = ", round( score_train ,4)) 
print (" in test MSE = ", score_test )
 
```

```python
# retransforming the predicted data which was one-hot encoded array

pred1 = pred.predict (x_test) 
pred1 = scaler_y.inverse_transform (np. array (pred1). reshape ((len( pred1), 1)))
```
