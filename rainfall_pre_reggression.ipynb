```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)
df=dataset1.groupby('SUBDIVISION').mean().sort_values(by='JAN')['ANNUAL']
df.plot('bar',width=0.3,title=' JAN AVG rainfall', fontsize=20,COLOR='b')
plt.xticks()
plt.ylabel('Average Rainfall')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)
df=dataset1.groupby('SUBDIVISION').mean().sort_values(by='ANNUAL')['ANNUAL']
df.plot('bar',width=0.3,title=' AVERAGE rainfall', fontsize=20)
plt.xticks()
plt.ylabel('Average Annual Rainfall ')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
```

```python
fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(111)
dfg = dataset1.groupby('YEAR').sum()['ANNUAL']
dfg.plot('line', title='Overall Rainfall in Each Year', fontsize=20,color='g')
plt.ylabel('Overall Rainfall ')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)

```

```python
months = dataset1.columns[15:]
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(111)
dataset1.groupby('YEAR').sum()[months].plot.line(title='Overall Rainfall ', ax=ax,fontsize=20)
plt.xticks(  rotation = 90)
plt.xlabel('SUBDIVISON')
plt.ylabel('Rainfall (mm)')
plt.legend(loc='upper right', fontsize = 'x-large')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
```

```python
months = dataset1.columns[2:14]
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(111)
dataset1.groupby('YEAR').sum()[months].plot.line(title='Overall Rainfall in Each Month of Year', ax=ax,fontsize=20)
plt.xticks(  rotation = 90)
plt.xlabel('SUBDIVISON')
plt.ylabel('Rainfall (mm)')
plt.legend(loc='upper right', fontsize = 'x-large')
ax.title.set_fontsize(30)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
```

```python

from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
groups = dataset1.groupby('SUBDIVISION')['JAN','SUBDIVISION','YEAR','JAN']
data=groups.get_group(('BIHAR'))
data
```

```python
groups
```

```python
cols=['YEAR']
```

```python
data.columns
```

```python

y = data['JAN']
x=data[cols]
```

```python
reg=linear_model.LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python

reg.fit(x_train,y_train)
```

```python

a=reg.predict(x_test)
a
```

```python
from sklearn.metrics import mean_squared_error
print('mse:%.2f'%mean_squared_error(y_test,a))
```
